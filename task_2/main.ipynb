```python
#Нахождение среднего бала из таблицы
import pandas as pd
students="data\students.xlsx"
df = pd.read_excel(students)
avg_math = round(df["Математика"].mean(),2)
avg_phiz = round(df["Физика"].mean(),2)
avg_inform = round(df["Информатика"].mean(),2)

```

```python
#Создание графика
import matplotlib.pyplot as plt
y=[avg_math,avg_phiz,avg_inform]
x=["Математика","Физика","Информатика"]
colors=['red','blue','green']
plt.bar(x, y, color=colors)
plt.xlabel('Прндмет')
plt.ylabel('Бал')
plt.legend()
plt.show()
```

```python
#Запись данных в новую таблицу
from openpyxl import load_workbook
file_path = "data\students_result.xlsx"
wb = load_workbook(file_path)
ws = wb.active

rows = [
    ["Математика", avg_math],
    ["Физика", avg_phiz],
    ["Информатика", avg_inform]
]

for r in rows:
    ws.append(r)

wb.save(file_path)
```
